```python
import matplotlib.pyplot as plt

from street_signs_identify.utils.x_image_loader import XImageLoader
from street_signs_identify.detector.craft_detector import CRAFTDetector
```

```python
data_path = "./dataset/2Penghu"
penghu_x = XImageLoader(data_path, annotation_file="x_labels.csv")

detector = CRAFTDetector()
```

```python
penghu_x._image_names
```

```python
image_gt = penghu_x["220720_231523537_Camera_0.jpg"]
plt.imshow(image_gt.draw_all_box())
image_gt.info
```

```python
detected_img = detector(image_gt.image)
detected_img.info
```

```python
plt.imshow(detected_img.draw_all_box())
```

```python
for detected_instance in detected_img:
    fig = plt.figure()
    ax = fig.add_subplot()
    ax.imshow(detected_instance.image)
    ax.set_title(detected_instance.info)
```
